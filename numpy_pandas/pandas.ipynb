```python
dane = pd.read_csv("titanic.csv")
```

```python
dane.info()
```

```python
dane["Survived"] = dane["Survived"].astype(bool)
```

```python
dane["Sex"] = dane["Sex"].astype("category")
```

```python
dane.info()
```

```python
# dane["Cabin"]
dane.Cabin.fillna("Deck", inplace=True)
```

```python
dane
```

### 1. Uzupełnij brakujące wartości w kolumnie "Age" średnią z istniejących wartości

```python
dane.Age.fillna(dane.Age.mean(), inplace=True)
dane
```

### 2. Uzupełnij brakujące wartości w kolumnie "Embarked" najpopularniejszą wartością z tejże kolumny

```python
k = dane.Embarked.mode()[0]
# k = dane.Embarked.value_counts().idxmax()
dane.Embarked.fillna(k, inplace=True)
```

### Znajdź wszystkich pasażerów, którzy zapłacili za bilet więcej niż 50 i nie byłi w 1. klasie

### Wypisz pasażerów, których nazwisko zawiera "Johnson"

```python
johnsons = dane[dane.Name.str.contains("Johnson")]
```

```python
johnsons.reset_index(drop=True).to_csv("johnsons.csv")
```
