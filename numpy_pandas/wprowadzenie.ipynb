## Broadcasting

Operacje możemy wykonywać tylko na tablicach kompatybilnych kształtów.
Aby sprawdzić kompatybilność 2 tablic wykonywane są następujące kroki:
- Tablica, która ma mniej wymiarów ma z lewej strony dostawiane nowe wymiary wielkości 1 (np. (2,3) -> (1,2,3)).
- Tablice są kompatybilne jeśli na odpowiadających sobie wymiarach mają tę samą wartość lub jedna z nich ma 1. Jeśli jedna z nich ma jeden to ta tablica jest powielana odpowiednią ilość razy wzdłuż tego wymiaru.

`(2, 3)` i `(2, 3)` - OK

`(4, 5)` i `(2, 4)` - Błąd

`(2,)` i `(3, 2)` - OK - `(2,)` jest rozszerzane do `(1, 2)`, a następnie jest powielane wzłuż 1. wymiaru do `(3, 2)`

`(3,)` i `(20, 10, 3)` - OK

````python
# (3, 2)
# [[1, 2],
#  [3, 4],
#  [5, 6]]

# (3, 2)
# [[10, 20],
#  [10, 20],
#  [10, 20]]

# -----------------

# (3, 2)
# [[1, 2],
#  [3, 4],
#  [5, 6]]

# (3, 2)
# [[10, 10],
#  [20, 20],
#  [30, 30]]
````
